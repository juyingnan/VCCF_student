# Test.PNG

```python
cell = cv2.imread('./Mask_RCNN/cell_image/test.png')
cell = cv2.cvtColor(cell, cv2.COLOR_BGR2RGB)
plt.imshow(cell)
```

```python
print("Cell Shape:",cell.shape)
```

```python
# Resizing

cell_resize3 = cv2.resize(cell, (int(cell.shape[1]*3), int(cell.shape[0]*3)))
plt.imshow(cell_resize3)
cell_resize3 = cv2.cvtColor(cell_resize3, cv2.COLOR_BGR2RGB)
cv2.imwrite("./Mask_RCNN/cell_image/test3.png",cell_resize3)
```

```python
# Create tiles

a = 0
for i in range(0,cell_resize3.shape[0],1080):
    for j in range(0,cell_resize3.shape[1],1920):
        tile = cell_resize3[i:i+1080,j:j+1920,:]
        if i+1080 <= cell_resize3.shape[0] and j+1920 <= cell_resize3.shape[1]:
            cv2.imwrite("./Mask_RCNN/glom_tiles/"+str(a)+".jpg",tile)
            a += 1
```
